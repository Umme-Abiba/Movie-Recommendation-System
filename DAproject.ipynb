```python
import ast

def parse_literal_eval(text):
    if isinstance(text, (list, dict)):
        return text
    try:
        return ast.literal_eval(text)
    except (ValueError, SyntaxError):
        return []

df['genres'] = df['genres'].apply(parse_literal_eval)
print("The 'genres' column has been parsed.")
```

## Calculate Median Budget and Average Vote Average per Genre



```python
genres_df = df.explode('genres')
print("DataFrame 'genres_df' created by exploding the 'genres' column.")
```

```python
genre_metrics = genres_df.groupby('genres').agg(
    median_budget=('budget', 'median'),
    average_vote_average=('vote_average', 'mean')
).reset_index()
print("Calculated median budget and average vote average for each genre.")
print(genre_metrics.head())
```

## Identify Overproduced Genres




```python
budget_threshold = genre_metrics['median_budget'].quantile(0.75)
vote_average_threshold = genre_metrics['average_vote_average'].quantile(0.25)

print(f"75th percentile of Median Budget: {budget_threshold}")
print(f"25th percentile of Average Vote Average: {vote_average_threshold}")
```

```python
overproduced_genres = genre_metrics[
    (genre_metrics['median_budget'] >= budget_threshold) &
    (genre_metrics['average_vote_average'] <= vote_average_threshold)
]
print("Identified Overproduced Genres:")
print(overproduced_genres)
```

```python
vote_average_threshold = genre_metrics['average_vote_average'].quantile(0.50)

print(f"75th percentile of Median Budget: {budget_threshold}")
print(f"50th percentile of Average Vote Average: {vote_average_threshold}")
```

```python
overproduced_genres = genre_metrics[
    (genre_metrics['median_budget'] >= budget_threshold) &
    (genre_metrics['average_vote_average'] <= vote_average_threshold)
]
print("Identified Overproduced Genres with adjusted vote average threshold:")
print(overproduced_genres)
```

```python
plt.figure(figsize=(12, 8))
sns.scatterplot(x='median_budget', y='average_vote_average', data=genre_metrics, hue='genres', s=100, alpha=0.7)

# Highlight overproduced genres
sns.scatterplot(x='median_budget', y='average_vote_average', data=overproduced_genres, color='red', s=200, marker='X', label='Overproduced Genres')

plt.title('Median Budget vs. Average Vote Average per Genre', fontsize=16)
plt.xlabel('Median Budget (USD)', fontsize=12)
plt.ylabel('Average Vote Average', fontsize=12)
plt.xscale('log') # Use log scale for budget for better visualization
plt.grid(True, which="both", ls="--", c='0.7')
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()
print("Scatter plot showing Median Budget vs. Average Vote Average per Genre displayed, with overproduced genres highlighted.")
```

As the budget increases beyond this initial range, the average ROI generally decreases or stabilizes at a much lower level compared to the lowest budget bin, even as gross profit continues to rise. This indicates that while higher budgets can lead to higher overall profits, the *efficiency* of the investment (ROI) might decrease.

## Analyze The "Flop" Profile Findings

### Characteristics of 'Flop' Movies
By comparing the descriptive statistics of 'runtime' and 'vote_average' for the bottom 10% of movies by gross profit (flop movies) against the entire dataset, we can identify key characteristics of movies that tend to underperform financially.

*   **Runtime**: The mean runtime for 'flop' movies is approximately 110.93 minutes, which is slightly higher than the overall mean runtime of 109.40 minutes for all movies. The standard deviation for flop movies' runtime is 23.99 minutes, compared to 21.43 minutes for all movies, suggesting a slightly wider spread in runtime for flop movies. This indicates that flop movies are not necessarily shorter or longer, but their runtime distribution might be slightly different.

*   **Vote Average**: The average vote average for 'flop' movies is 5.69, which is noticeably lower than the overall average vote average of 6.30 for all movies. The 25th percentile for flop movies is 5.1, the median is 5.7, and the 75th percentile is 6.3. In contrast, for all movies, these values are 5.8, 6.4, and 6.9, respectively. This clearly suggests that movies with lower critical or audience reception (as indicated by 'vote_average') are more likely to be financial 'flops'.

**Conclusion**: 'Flop' movies, as defined by being in the bottom 10% of gross profit, tend to have a significantly lower average vote average compared to the general movie population. Their runtime distribution is similar to the overall dataset, suggesting that runtime alone is not a strong indicator of a movie being a flop, but audience reception is.
